### importing

```python
import pandas as pd 
```

```python
train = pd.read_csv('train.csv')

train.head()
```

```python
train.info()
```

### manipulation 

```python
train.Sex.unique()
```

```python
train.Embarked.unique()
```

```python
sex_uni = train.Sex.unique().tolist()
train.Sex = train.Sex.map(lambda x : sex_uni.index(x))

Embarked_uni = train.Embarked.unique().tolist()
train.Embarked = train.Embarked.map(lambda x : Embarked_uni.index(x))

train.drop(['Name','Ticket','Cabin','Age'],axis=1,inplace=True)
```

```python
train.info()
```

```python
train_sol = train.Survived
train.drop('Survived',axis=1,inplace=True)
```

```python
train.info()
```

### shuffling the data 

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(train,train_sol,test_size=0.2)
```

```python
test_x
```

### using a model 

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

clf.fit(train_x,train_y)

model_result = clf.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

accuracy_score(test_y,model_result)
```

```python
confusion_matrix(test_y,model_result)
```

```python
test = pd.read_csv('test.csv')

sex_uni = test.Sex.unique().tolist()
test.Sex = test.Sex.map(lambda x : sex_uni.index(x))

Embarked_uni = test.Embarked.unique().tolist()
test.Embarked = test.Embarked.map(lambda x : Embarked_uni.index(x))

test.drop(['Name','Ticket','Cabin','Age'],axis=1,inplace=True)

test.Fare.fillna(test.Fare.mean(),inplace=True)
```

```python
clf = RandomForestClassifier()

clf.fit(train,train_sol)
result = clf.predict(test)
```

```python
df = pd.DataFrame({'PassengerId':test.PassengerId,'Survived':result})
```

```python
df.set_index('PassengerId').to_csv('result.csv')
```
